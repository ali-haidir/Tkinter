```python
from tkinter import *
```

```python
root = Tk()
root.title("Simple Calculater")

def button_add(number):
    current =entry_field.get()
    entry_field.delete(0,END)
    entry_field.insert(0, current + str(number))
    
def button_cleear():
    entry_field.delete(0, END)
    
def button_add_fn():
    global f_num, sign
    f_num = int(entry_field.get())
    sign = "+"
    entry_field.delete(0,END)

    
def button_subtract_fn():
    global f_num, sign
    f_num = int(entry_field.get())
    sign = "-"
    entry_field.delete(0,END)

    

def button_multiply_fn():
    global f_num, sign
    f_num = int(entry_field.get())
    sign = "*"
    entry_field.delete(0,END)



def button_devide_fn():
    global f_num, sign
    f_num = int(entry_field.get())
    sign = "/"
    entry_field.delete(0,END)

    
    
def button_equal_fn():
    s_num = int(entry_field.get())
    entry_field.delete(0,END)
   
    if sign == "+":
        answer = f_num +s_num
    if sign == "-":
        answer = f_num -s_num
    if sign == "*":
        answer = f_num *s_num
    
    if sign == "/":
        answer = f_num / s_num
        

    entry_field.insert(0, answer)
    
    
    

entry_field = Entry(root , width= 35 , borderwidth = 3)
entry_field.grid(row = 0 , column =0 , columnspan = 3 , padx = 10 ,pady = 10)

button1 = Button( root , text = "1" , padx = 40 ,pady = 40 , command = lambda: button_add(1)  )
button1.grid(row =1 , column = 0)

button2 = Button( root , text = "2" , padx = 40 ,pady = 40 , command = lambda :button_add(2)  )
button2.grid(row =1 , column = 1)

button3 = Button( root , text = "3" , padx = 40 ,pady = 40 , command = lambda :button_add(3)  )
button3.grid(row =1 , column = 2)

button4 = Button( root , text = "4" , padx = 40 ,pady = 40 , command = lambda :button_add(4)  )
button4.grid(row =2 , column = 0)

button5 = Button( root , text = "5" , padx = 40 ,pady = 40 , command = lambda :button_add(5)  )
button5.grid(row =2 , column = 1)

button6 = Button( root , text = "6" , padx = 40 ,pady = 40 , command =lambda :button_add(6)  )
button6.grid(row =2 , column = 2)

button7 = Button( root , text = "7" , padx = 40 ,pady = 40 , command =lambda :button_add(7)  )
button7.grid(row =3 , column = 0)

button8 = Button( root , text = "8" , padx = 40 ,pady = 40 , command =lambda :button_add(8)  )
button8.grid(row =3 , column = 1)

button9 = Button( root , text = "9" , padx = 40 ,pady = 40 , command =lambda :button_add(9))
button9.grid(row =3 , column = 2)

button10 = Button( root , text = "0" , padx = 40 ,pady = 40 , command =lambda :button_add(0)  )
button10.grid(row =4 , column = 1)


button_addd = Button( root , text = "+" , padx = 40 ,pady = 40 , command =lambda :button_add_fn()  )
button_equalto = Button( root , text = "=" , padx = 40 ,pady = 40 , command =lambda :button_equal_fn())
button_clear = Button( root , text = "clear the screen " , padx = 20 ,pady = 20 , command =lambda :button_cleear())
button_subtract = Button( root , text = "-" , padx = 40 ,pady = 40 , command =lambda :button_subtract_fn())
button_multiply = Button( root , text = "*" , padx = 40 ,pady = 40 , command =lambda :button_multiply_fn())
button_devide = Button( root , text = "/" , padx = 40 ,pady = 40 , command =lambda :button_devide_fn())



button_addd.grid(row = 4 , column = 0)
button_equalto.grid(row = 4 , column = 2)
button_clear.grid(row = 6 , column = 0 , columnspan = 3)
button_subtract.grid(row = 5 , column = 0)
button_multiply.grid(row = 5 , column = 1)
button_devide.grid(row = 5 , column = 2)


root.mainloop()
```
