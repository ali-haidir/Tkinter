```python
from tkinter import *
import PIL 
from PIL import *
from PIL import ImageTk
from PIL import Image
```

```python
root = Tk()
root.title("icon & exit button ")
#root.iconbitmap("/home/hydir/Pictures/triangle.png")



my_img = ImageTk.PhotoImage(Image.open("triangle.png"))
my_img1 = ImageTk.PhotoImage(Image.open("image.png"))

   
global image_number 
image_number = 0

image_list = [my_img , my_img1]


my_label = Label(image = image_list[image_number])
my_label.grid(row = 0 , column = 0 ,columnspan =3)





def next_button_fn(image_number):
    
    
    my_label = Label(image = image_list[image_number])
    
    my_label.grid_forget()
    
   

   # print(len(image_list))
    if image_number >= len(image_list ) -1 :
        next_button = Button(root , text=">>",  padx =20 , pady = 20,  state = DISABLED  )
        
    else:
        
        image_number +=1
        count  = image_number
        
        my_label = Label(image = image_list[image_number])
        my_label.grid(row = 0 , column = 0 ,columnspan =3)
        
        
        
     
    status = Label (root , text = "page "+ str(image_number +1) + " of " + str(len(image_list)) ,bd = 2 , relief = SUNKEN , anchor = E)
    status.grid(row = 2 , column = 0 , columnspan = 3,sticky = W+E ,  )
    return count 

def pre_button_fn(image_number):
        print(image_number)
        if image_number <= 0 :
            pre_button = Button(root , text=">>",  padx =20 , pady = 20,  state = DISABLED  )
            

        else:
            
            image_number -=1
            print(image_number)
        
            my_label = Label(image = image_list[image_number])
            my_label.grid(row = 0 , column = 0 ,columnspan =3)




exit_button = Button(root, text="exit" ,command = root.destroy,  padx =20 , pady = 20)
next_button = Button(root , text=">>",  padx =20 , pady = 20, command = lambda: next_button_fn(image_number))
pre_button = Button(root , text = "<<",  padx =20 , pady = 20,command = lambda: pre_button_fn(next_button))



exit_button.grid(row = 1  , column = 1,  )
next_button.grid(row = 1  , column = 2,  )
pre_button.grid(row = 1  , column = 0, )

  
status = Label (root , text = "page "+ str(image_number +1) + " of " + str(len(image_list)) ,bd = 2 , relief = SUNKEN , anchor = E)
status.grid(row = 2 , column = 0 , columnspan = 3,sticky = W+E ,  )



root.mainloop()
```
